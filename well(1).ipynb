```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Well dimensions
diameter1 = 2.80  # m
diameter2= 1.0  # m
a = diameter1/2
b = diameter2/2  # m
depth = -7  # Depth in meters

# Generate the parameterization
theta = np.linspace(0, 2 * np.pi, 100)  # Angle
z = np.linspace(0, depth, 50)  # Depth
theta_grid, z_grid = np.meshgrid(theta, z)

# coord
x = b * np.cos(theta_grid)
y = a * np.sin(theta_grid)

# plot
fig = plt.figure(figsize=(20, 18))  
ax = fig.add_subplot(111, projection='3d')


# surface 
ax.plot_surface(
    x, y, z_grid,
    facecolors=plt.cm.coolwarm((z_grid - z_grid.min()) / (z_grid.max() - z_grid.min())),
    edgecolor='none',
    alpha=0.8
)

# aspect ratio 
ax.set_box_aspect([b, 2*a/3, abs(depth) / 10])  

ax.set_title("Visualization of the Well", fontsize=24)  
ax.set_xlabel("Width in meters (X-axis)", labelpad=25, fontsize=18, ) 
ax.set_ylabel("Length in meters (Y-axis)", labelpad=35 , fontsize=18)  
ax.set_zlabel("Depth in meters (Z-axis)", labelpad=15, fontsize=18) 

# tick label sizes
ax.tick_params(axis='x', labelsize=12, pad=15)  
ax.tick_params(axis='y', labelsize=12, pad=15) 
ax.tick_params(axis='z', labelsize=12, pad=15) 

# Tilt the x-axis labels
for label in ax.get_xticklabels():
    label
    label.set_rotation(-30)  
    label.set_horizontalalignment('right') 

# Tilt the y-axis labels
for label in ax.get_yticklabels():
    label.set_rotation(30) 
    label.set_horizontalalignment('left') 


# declutter this shit
a= a+0.5
b= b+1
ax.set_xticks(np.linspace(-b, b, 6))
ax.set_yticks(np.linspace(-a, a, 10))

ax.view_init(elev=30, azim=45)

output_file = "elliptical_cylinder_visualization.png"
if os.path.exists(output_file):
    os.remove(output_file)

plt.savefig('elliptical_cylinder_visualization.png', bbox_inches='tight')
print(f"Plot saved as {output_file}")

plt.show()
```
